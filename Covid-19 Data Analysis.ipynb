# Relating Confirmed Cases with Cured with a Line Plot

This shows us that the number of people getting cured from corona is proportional to the confirmed cases as the number of cases is rising, the number of patients getting cured is also increasing proportionately.

```python
sns.relplot(x="Confirmed", y = "Cured", kind ="line", data = data)
```

# Relation of Deaths with date using Line plot

By This Final Visualization of "Deaths with date" we can notice that in the initial month, the graph seems to be flat which means less number of deaths, but as the months goes by, we see the peak of the 1st Wave of Covid-19 around  September 2020, but after that we can see the the death rates to be constant for sometime but still high, Until We Get Hit by another wave of Covid-19 with Peak in May 2021, and It was as close as 10,000 deaths per day.

```python
sns.relplot(x="Date", y = "Deaths", kind ="line", data = data)
```
